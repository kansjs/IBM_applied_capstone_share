```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# Extent of total area and setup of grids
North_max = 39.27
South_max = 38.69
East_max = -76.71
West_max = -77.45
NS_axis = 7
EW_axis = 6
```

```python
NS_diff = North_max - South_max
EW_diff = East_max - West_max
```

```python
column_names = ['Lat1', 'Long1', 'Lat2', 'Long2'] 
DMV_grid = pd.DataFrame(columns=column_names)
DMV_grid
```

```python
# Creation of grid
for f in range(NS_axis):
    for g in range(EW_axis):
        temp_lat1 = North_max - f * (NS_diff/NS_axis)
        temp_lat2 = North_max - (f + 1) * (NS_diff/NS_axis)
        temp_long1 = East_max - g * (EW_diff/EW_axis)
        temp_long2 = East_max - (g + 1) * (EW_diff/EW_axis)
        temp = pd.DataFrame([temp_lat1, temp_long1, temp_lat2, temp_long2]).T
        temp.columns = column_names
        DMV_grid = DMV_grid.append(temp)
DMV_grid
```
